```python
from seagal import seagal
```

```python
sg = seagal.SEAGAL(visium_path='data/VisiumData/Human_Breast_Cancer_BAS1')
```

```python
gene_groups = {"T": ["Il7r", "Cd3d", "Cd3e", "Cd3f", "Itk", "Lef1", "Skap1"],
               'B': ["Igkc", "Cd79b", "Cd79a", "Ebf1", "Bank1", "Ighd"],
               'Neutrophils': ["Mmp9", "S100a8", "S100a9"]}

seagal.group_adata_by_genes(sg, gene_groups, inplace=True)
```

```python
seagal.clustermap(sg, figsize=(3,3), use_grouped=True)
```

```python
f,_= seagal.hotspot(sg, 'B', 'T', use_grouped=True, cmap='Reds')
```
